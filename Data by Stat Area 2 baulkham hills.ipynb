```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
baulkham_hills_df = pd.read_csv(Path("Resource/Data by Stat Area 2 baulkham hills.csv"))
baulkham_hills_df = baulkham_hills_df.iloc[0:28]
baulkham_hills_df.head(30)
```

```python
baulkham_hills_df = baulkham_hills_df.iloc[[11,25]]
baulkham_hills_df.drop(columns=['Australia', 'Description', 'Year'], inplace=True)
baulkham_hills_mortgage = baulkham_hills_df
baulkham_hills_mortgage
```

```python
baulkham_hills_east = baulkham_hills_df.loc[baulkham_hills_df['Statistical Area Level 2 (SA4)'] == 'Baulkham Hills - East']
baulkham_hills_east.head()
```

```python
baulkham_hills_west = baulkham_hills_df.loc[baulkham_hills_df['Statistical Area Level 2 (SA4)'] == 'Baulkham Hills (West) - Bella Vista']
baulkham_hills_west.head()
```

```python
baulkham_hills_mortgage['Region'] = baulkham_hills_mortgage['Region'].astype(int)
baulkham_hills_mortgage.set_index('Statistical Area Level 2 (SA4)', inplace=True)
baulkham_hills_mortgage_bar = baulkham_hills_mortgage['Region'].hvplot.bar(label='Median Monthly Mortgage')
baulkham_hills_mortgage_bar
```
