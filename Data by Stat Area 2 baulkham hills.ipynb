```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
baulkham_hills_df = pd.read_csv(Path("Resource/Data by Stat Area 2 baulkham hills.csv"))
baulkham_hills_df.head(500)
```

```python
baulkham_hills_df = baulkham_hills_df.iloc[[11,25,39,53,67,81,95,109,123,137,151,165,179,193,207,221,235,249,263,277]]
baulkham_hills_df.drop(columns=['Australia'], inplace=True)
baulkham_hills_mortgage = baulkham_hills_df
baulkham_hills_mortgage
```

```python
baulkham_hills_mortgage['Region'] = baulkham_hills_mortgage['Region'].astype(int)
baulkham_hills_mortgage.set_index('Statistical Area Level 2 (SA4)', inplace=True)
baulkham_hills_mortgage_bar = baulkham_hills_mortgage['Region'].hvplot.bar(
    label='Median Monthly Mortgage of Each Suburb in Baulkham Hills',
    ylabel='Median Income',
    rot=60, 
    height = 500
)
baulkham_hills_mortgage_bar
```

```python
baulkham_hills_mortgage_scatter = baulkham_hills_mortgage['Region'].hvplot.scatter(
    label='Median Monthly Mortgage Plot of Each Baulkham Hill Suburb',
    rot=60, 
    height = 500
)
baulkham_hills_mortgage_scatter
```

```python
baulkham_hills_mortgage_box = baulkham_hills_mortgage['Region'].hvplot.box(
    label='Median Monthly Mortgage of Baulkham Hills',
    y="Region",
    rot=30, 
    height = 800,
    width = 400
)
baulkham_hills_mortgage_box
```
