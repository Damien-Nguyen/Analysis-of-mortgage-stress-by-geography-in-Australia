```python
mortgages_stress_2021 = NSW_mortgages_DF.drop(columns=["Parent Description","Description","2016","2017","2018","2019","2020"])
mortgages_stress_2021.drop([0,1,244,245,414,455,642,643,646,647,742,743,744,745,746,747,748,749], axis=0, inplace=True)
Lower_income = mortgages_stress_2021.drop(index=[641,645])
Lower_income_no = mortgages_stress_2021.drop(index=[641,644,645])
Lower_income_percentage = mortgages_stress_2021.drop(index=[640,641,645])
Higher_income = mortgages_stress_2021.drop(index=[640,644])
Higher_income_no = mortgages_stress_2021.drop(index=[640,644,645])
Higher_income_percentage = mortgages_stress_2021.drop(index=[640,641,644])
Lower_vs_higher_no = mortgages_stress_2021.drop(index=[644,645])
Lower_vs_higher_percentage = mortgages_stress_2021.drop(index=[640,641])
mortgages_stress_2021.set_index(mortgages_stress_2021['Measure Code'], inplace=True)
mortgages_stress_2021.drop(columns=['Measure Code'], inplace=True)
mortgages_stress_2021
```

```python
Lower_vs_higher_no.set_index(Lower_vs_higher_no['Measure Code'], inplace=True)
Lower_vs_higher_no.drop(columns=['Measure Code'], inplace=True)
Lower_vs_higher_no_plot = Lower_vs_higher_no['2021'].hvplot.bar(
    label="Lower VS Higher Mortgage income 2021 (no.Households)",
    xlabel='Income Rates',
    ylabel = 'No. of Households',
    shared_axes = False,
    width = 600)
Lower_vs_higher_no_plot
```

```python
Lower_vs_higher_percentage.set_index(Lower_vs_higher_percentage['Measure Code'], inplace=True)
Lower_vs_higher_percentage.drop(columns=['Measure Code'], inplace=True)
Lower_vs_higher_percentage_plot = Lower_vs_higher_percentage['2021'].hvplot.bar(
    x='Measure Code',
    xlabel='(%) Income Rates',
    ylabel = '(%) No. of Households',
    label="Lower VS Higher Mortgage income (%) 2021",
    shared_axes = False,
    width = 600,
).opts(color = 'orange'
)
Lower_vs_higher_percentage_plot
```

```python
Lower_vs_higher_percentage.plot(
    kind='pie',
    subplots=True,
    title = 'Lower vs Higher Household Mortgage Income 2021(%)',
    figsize=(8,8))
plt.show
```
